```python
# OS
import os

# environment library
from dotenv import load_dotenv
load_dotenv()

# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine

# mathplotlib
import matplotlib.pyplot as plt

```

```python
# Create a connection to the postgresl database
database_url = "postgresql://" + os.environ.get('DataBase')

engine = create_engine(database_url)
conn = engine.connect()
```

```python
# Query Records in the the Employees Table and connect supporting tables.
#employee_data = pd.read_sql("SELECT * FROM employees",conn)

sql_cmd = "select e.emp_no, e.last_name, e.first_name, titles.title, s.salary "
sql_cmd = sql_cmd + "from employees as e "
sql_cmd = sql_cmd + "inner join titles on e.emp_title_id=titles.title_id "
sql_cmd = sql_cmd + "inner join salaries as s on e.emp_no=s.emp_no "
#print(sql_cmd)

employee_data = pd.read_sql(sql_cmd,conn)

employee_data.info()
```

```python
# Histogram of Salary Data for Pewlett Hackard from the data in the Postgresql DB

# Gather the data needed for the histogram
salary_data = employee_data["salary"]

# Create the histogram chart.
plt.figure(figsize = (7,6))
plt.hist(salary_data)

# decorate the histogram chart for easier reading.
plt.title('Histogram of Salary Data for Pewlett Hackard')
plt.ylabel('Employee Counts')
plt.xlabel('Salary Range')

# Display the bar chart.
plt.tight_layout()
plt.show()
```

```python
# Average Salaries by Employee Tittle for Pewlett Hackard bar chart.

# Needs some help from numpy
import numpy as np

#Group the data by Employee Title and calculate the mean of the groups.
salary_by_dept = employee_data.groupby(["title"]).agg(avg_salary = ('salary', 'mean'))
                                                                 

# Gather the X and Y data for the bar chart.
x_labels = salary_by_dept.index
avg_salary = salary_by_dept["avg_salary"]

#Set the X axis for bar chart formatting.
x_axis = np.arange(len(x_labels))
tick_locations = [value + .01 for value in x_axis]

# Create the bar chart. 
plt.figure(figsize = (7,6))
plt.bar(x_axis, avg_salary, align="center", alpha= 0.5)
plt.xticks(tick_locations, x_labels, rotation="vertical")
plt.xlim(-0.50, len(x_axis))

# decorate the bar chart for easier reading.
plt.title("Average Salaries by Employee Titles for Pewlett Hackard")
plt.xlabel("Employee Title Names")
plt.ylabel("Average Salaries")

# Display the bar chart.
plt.tight_layout()
plt.show()
```
